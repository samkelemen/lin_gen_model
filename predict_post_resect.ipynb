```python
import data_manager
import matplotlib.pyplot as plt
import seaborn as sns


pre_resect_rules_gl = data_manager.get_secondary_data(f"pre_resection/gl/matrices/fitted_O/fitted_O")

subject_ids = [1, 2, 3, 5, 6, 7, 8, 10, 11, 13, 14, 15, 16, 17, 19, 20, 22, 23, 24, 25, 26, 27, 28, 29, 31]
for subject_id in subject_ids:
    pre_sc_matrix = data_manager.get_pre_resection_sc(subject_id)
    post_sc_matrix = data_manager.get_post_resection_sc(subject_id)
    post_fc_matrix = data_manager.get_post_resection_fc(subject_id)
    pre_fc_matrix = data_manager.get_pre_resection_fc(subject_id)
    pre_resect_rules = data_manager.get_secondary_data(f"pre_resection/sl/matrices/fitted_O/fitted_O{subject_id}")

    prediction_post = post_sc_matrix @ pre_resect_rules @ post_sc_matrix
    prediction_pre = pre_sc_matrix @ pre_resect_rules @ pre_sc_matrix
    prediction_post_gl = post_sc_matrix @ pre_resect_rules_gl @ post_sc_matrix

    sns.heatmap(pre_sc_matrix, cmap='Greys', center=0)
    plt.title("SC Pre Op")
    plt.show()
    plt.close()

    sns.heatmap(post_sc_matrix, cmap='Greys', center=0)
    plt.title("SC Post Op")
    plt.show()
    plt.close()

    sns.heatmap(prediction_pre, cmap='vlag', center=0)
    plt.title("Predicted FC Pre Op")
    plt.show()
    plt.close()

    sns.heatmap(prediction_post_gl, cmap='vlag', center=0)
    plt.title("Predicted FC Post Op GL")
    plt.show()
    plt.close()

    sns.heatmap(pre_fc_matrix, cmap='vlag', center=0)
    plt.title("Actual FC Pre Op")
    plt.show()
    plt.close()

    sns.heatmap(prediction_post, cmap='vlag', center=0)
    plt.title("Predicted FC Post Op")
    plt.show()
    plt.close()

    sns.heatmap(post_fc_matrix, cmap='vlag', center=0)
    plt.title("Actual FC Post Op")
    plt.show()
    plt.close()

    
```
