```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt


TUMOR_IDS = [1, 2, 3, 5, 6, 7, 8, 10, 13, 15, 16, 17, 20, 24, 25, 26, 28]

gl_path = "C:/Users/samke/OneDrive - University of Kansas/Documents1/Computational Neurocognition Lab/LinGenModel/lin_gen_model/pre_resec_train_output/gl/matrices/fitted_O/fitted_O"
sl_path = "C:/Users/samke/OneDrive - University of Kansas/Documents1/Computational Neurocognition Lab/LinGenModel/lin_gen_model/pre_resec_train_output/sl/matrices/fitted_O/fitted_O"

group_rules = np.loadtxt(gl_path)

mean_rules = np.zeros((68, 68))
rules_stack = []
for subject_id in TUMOR_IDS:
    rules_stack.append(np.loadtxt(sl_path + f"{subject_id}"))
    mean_rules += np.loadtxt(sl_path + f"{subject_id}")
mean_rules /= len(TUMOR_IDS)

sns.heatmap(group_rules, cmap="coolwarm", center=0)
plt.show()

sns.heatmap(mean_rules, cmap="coolwarm", center=0)
plt.show()

sns.heatmap(mean_rules - group_rules, cmap="coolwarm", center=0)
plt.show()

sns.heatmap(np.var(np.array(rules_stack), axis=0), cmap="Greys", vmax=20000)
plt.show()
```
